**Normalization of data**

```python
def norm(i):
  x=(i-i.min()) / (i.max()-i.min())
  return(x)
```

```python
df_norm = norm(crime.iloc[:,1:])
df_norm
```

**Hierarchial Clustering**

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score as sil
```

**Average Linkage**

```python
plt.figure(figsize=(15,5))
dendogram=sch.dendrogram(sch.linkage(df_norm, method='average'))
```

```python
hc_avg=AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='average')
```

```python
y_hc_avg = hc_avg.fit_predict(df_norm)
y_hc_avg
```

```python
df_1=pd.DataFrame(y_hc_avg, columns=['hc_avg_clusters'])
```

```python
crime_hc_avg=pd.concat([crime,df_1], axis=1)
crime_hc_avg
```

```python
crime_hc_avg.groupby('hc_avg_clusters').mean().reset_index()
```

```python
score_avg = (sil(df_norm, hc_avg.labels_)*100).round(2)
score_avg
```

**Complete Linkage**

```python
plt.figure(figsize=(15,5))
dendogram=sch.dendrogram(sch.linkage(df_norm, method='complete'))
```

```python
hc_complete=AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='complete')
```

```python
y_hc_comp = hc_complete.fit_predict(df_norm)
y_hc_comp
```

```python
df_2=pd.DataFrame(y_hc_comp, columns=['hc_comp_clusters'])
```

```python
crime_hc_comp=pd.concat([crime,df_2], axis=1)
crime_hc_comp
```

```python
crime_hc_comp.groupby('hc_comp_clusters').mean().reset_index()
```

```python
score_comp = (sil(df_norm, hc_complete.labels_)*100).round(2)
score_comp
```

**Centroid Linkage**

```python
plt.figure(figsize=(15,5))
dendogram=sch.dendrogram(sch.linkage(df_norm, method='ward'))
```

```python
hc_centroid=AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')
```

```python
y_hc_cent = hc_centroid.fit_predict(df_norm)
y_hc_cent
```

```python
df_3=pd.DataFrame(y_hc_cent, columns=['hc_cent_clusters'])
```

```python
crime_hc_cent=pd.concat([crime,df_3], axis=1)
crime_hc_cent
```

```python
crime_hc_cent.groupby('hc_cent_clusters').mean().reset_index()
```

```python
score_cent = (sil(df_norm, hc_centroid.labels_)*100).round(2)
score_cent
```

K=Means

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)

print(wcss)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
K_clusters=KMeans(4, random_state=0)
K_clusters.fit(df_norm)
```

```python
KM_label=K_clusters.labels_
KM_label
```

```python
df_3=pd.DataFrame(KM_label, columns=['KMeans_labels'])
```

```python
y_KM=pd.concat([crime,df_3], axis=1)
y_KM
```

```python
y_KM.groupby('KMeans_labels').mean().reset_index()
```

```python
score_KM = (sil(df_norm, KM_label)*100).round(2)
score_KM
```

**DBSCAN**

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan=DBSCAN(eps=0.3, min_samples=3)
dbscan.fit(df_norm)
```

```python
dbscan_label=dbscan.labels_
dbscan_label
```

```python
df_4=pd.DataFrame(dbscan.labels_, columns=['Cluster_DBSCAN'])
```

```python
y_DBSCAN=pd.concat([crime, df_4],axis=1)
y_DBSCAN.head()
```

```python
score_DBSCAN = (sil(df_norm, dbscan.labels_)*100).round(2)
score_DBSCAN
```
